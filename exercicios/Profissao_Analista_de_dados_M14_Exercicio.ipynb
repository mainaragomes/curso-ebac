## 1\. Preço do diamante por tipo de corte

```
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("diamonds")
data.head()
```

```
# gráfico do exercício 1

diamonds = sns.barplot(data= data, x = "cut", y = "price", errorbar= None, palette = "dark")
diamonds.set_title("Valor do diamante por tipo de corte")
diamonds.set(xlabel="Tipo de corte", ylabel="Preço do diamante")
```

**Insight do gráfico 1**: Pelo gráfico, é possível perceber que, quanto mais refinado o corte do diamante, mais ele vale, como se nota no corte premium. O menor valor diz respeito ao corte ideal. É possível notar, também, que os cortes "good" e "very good" quase não possuem diferença de valor.

## 2\. Número de passageiros em dezembro por ano

```
import seaborn as sns
import pandas as pd

data = sns.load_dataset("flights")
data.head()
```

```
df = pd.DataFrame(data)

data_december = df[df['month'] == 'Dec']
print(data_december)
```

```
# gráfico do exercício 2

flights = sns.lineplot(data= data_december, x = "year", y = "passengers", color = "red")
flights.set_title("Número de passageiros em dezembro por ano")
flights.set(xlabel="Ano", ylabel="Número de passageiros")
```

**Insight do gráfico 2**: O gráfico acima nos mostra que o número de passageiros no mês de dezembro foi crescendo no decorrer dos anos, exceto nos anos 1953 e 1958 em que se nota uma queda.

## 3\. Numero de passageiros por mês entre 1949 e 1959

```
data = sns.load_dataset("flights")
data.head()
```

```
# gráfico do exercício 3
flights = data.query('1949 <= year <= 1959')
months_order = ['January', 'February', 'March', 'April', 'May', 'June',
                'July', 'August', 'September', 'October', 'November', 'December']

plt.figure(figsize=(10, 6))
with sns.axes_style('whitegrid'):
  flights_plot = sns.lineplot(data=flights, x='month', y='passengers', palette='bright', errorbar=None, hue='year')
  flights_plot.set_title("Número de passageiros por mês entre 1949 e 1959")
  flights_plot.set(xlabel="Mês", ylabel="Passageiros")
  flights_plot.legend(title="Ano")
  plt.xticks(rotation=45, ha='right')

plt.show()
```

**Insight do gráfico 3**: O gráfico 3 nos mostra que, entre 1949 e 1959, houve, de certa forma, uma constância no número de passageiros por mês, tanto nos meses em que há um crescimento, quanto nos meses em que há um decréscimo. Os meses em que mais houve passageiros durante o período, são julho e agosto. Nota-se que, em todos os anos, a partir de agosto o número de passageiros passa a cair, assim como nos meses de maio e dezembro começam a aumentar. Tais meses, no Brasil, são de férias escolares, o que pode ser uma das causas no aumento do movimento.
