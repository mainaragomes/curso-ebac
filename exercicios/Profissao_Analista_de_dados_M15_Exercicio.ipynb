## 1\. Preços `outliers` de diamante

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
# gráfico do exercício 1

with sns.axes_style("whitegrid"):
  diamonds = sns.boxplot(x = data['price'])
  diamonds.set_title("Preços de diamante")
  diamonds.set_xlabel("Preço")
```

**Insight do gráfico 1**: A partir do gráfico 1, é possível perceber que a maioria dos preços dos diamantes está entre 1000 e 5000. Quando os preços começam a ficar mais caros, a frequência diminui. Presume-se que quanto maior o valor, mais raro é o diamante.

## 2\. Correlação entre o do preço do diamante com seu peso (`carat`) agrupados por sua transparêcia (`clarity`)

```
# gráfico do exercício 2

diamonds = data[["price", "carat", "clarity"]]

with sns.axes_style("ticks"):

  grafico = sns.scatterplot(data=diamonds, x="price", y="carat", hue="clarity", palette="dark")
  grafico.set(title="Correlação entre o do preço do diamante com seu peso (carat) agrupados por sua transparêcia (clarity)", xlabel="Preço", ylabel="Peso")
  grafico.figure.set_size_inches(w=50/2.54, h=20/2.54)
  grafico.get_legend().set_title("Tranparência")
```

**Insight do gráfico 2**: De acordo com o gráfico 2, é notório que, quanto menor o peso e maior o preço, a probabilidade de ser das clarezas IF, WS1 e WS2, é maior. Por outro lado, quanto maior o peso, maior se torna o seu preço.

## 3\. Distribuição contínua aproximada do peso (`weight`) de carros

```
import seaborn as sns

data = sns.load_dataset("mpg")
data.head()
```

```
# gráfico do exercício 3

cars = data[["weight", "origin"]]

with sns.axes_style("white"):
  grafico = sns.histplot(data=cars, x="weight", hue="origin", kde=True, palette= 'dark')
  grafico.set(title="Distribuiçao do peso dos carros de acordo com a origem", xlabel="Peso", ylabel="Contagem");
  grafico.get_legend().set_title("Origem")
```

**Insight do gráfico 3**:  É possível perceber, de acordoc com o gráfico,que os carros japoneses são os mais leves. Os carros de origem europeia começam a ter uma decadência a partir de 2.500kg de peso, bem como os carros japoneses (os quais tem sua maior frequência em 2000kg). Além disso, é possível observar que os carros de origem dos Estados Unidos, começam a cair a partir de 4000kg. Sendo assim, é possível afirmar que os carros de origem estadunidense são os mais pesados.
